```python
import matplotlib.pyplot as plt
```

```python
sigma = [0.1, 0.2, 0.3]
mpor = [1, 2, 3]
alpha = [98, 99, 99.5]
```

```python
for c in alpha:
  for s in sigma:
    for m in mpor:
      a = (c-50)/21.063
      im = s * (m/252)**0.5 * a
      print('C.I: ', c, 'Vol: ', s, 'MPoR: ', m, 'IM: ', round(100*im, 2) )
```

```python
# create data
mpor = [1, 2, 3]
im_1 = [1.48, 2.09, 2.56]
im_2 = [2.96, 4.19, 5.13]
im_3 = [4.44, 6.28, 7.69]

# plot lines
plt.plot(mpor, im_1, label = '$\sigma = 0.1$')
plt.plot(mpor, im_2, label = '$\sigma = 0.2$')
plt.plot(mpor, im_3, label = '$\sigma = 0.3$')
plt.legend()
plt.xlabel('MPoR, days')
plt.xticks(mpor)
plt.ylabel('IM, %')
plt.title("IM at 99% confidence level")
plt.show()
```
